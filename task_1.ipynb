```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
url ="http://bit.ly/w-data"
data = pd.read_csv(url)
print("data download success ")
print(data.head(5))
```

## plotting tne distribution 

```python
# plotting the distribution of scores
data.plot(x='Hours',y='Scores',style='o')
plt.title('hours vs pourcentage')
plt.xlabel('Hours')
plt.ylabel('scores')
plt.show()
```

### From the graph above, we can clearly see that there is a positive linear relation between the number of hours studies and percentage of score.

## Prepartion of the data

```python
# divide the data into "attrubutes" (inputs) and "labels" (outputs) for the training model
X =data.iloc[:,:-1]
y = data.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

## training the algorithm

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
# plotting the regression line
line =regression.coef_*X+regression.intercept_

# plotting for the test data 
plt.scatter(X,y)
plt.plot(X,line)
plt.show()
```

## Making predictions

```python
print(X_test)
y_pred = regression.predict(X_test)
```

```python
# comparing actual vs predicted
df = pd.DataFrame({'Actual':y_test , 'Predicted':y_pred})
print(df)
```

```python
hours = 9.25
# Reformater 'hours' en un tableau 2D
hours_reshaped = np.array(hours).reshape(-1, 1)
# Faire la prédiction
own_pred = regression.predict(hours_reshaped)
#own_pred=regression.predict(hours)
print("No of hours = {}".format(hours))
print("Predicted score ={}".format(own_pred[0]))
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
